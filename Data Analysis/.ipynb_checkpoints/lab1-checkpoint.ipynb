## Предобработка

### Обработка данных в столбцах
Изучим данные из файла:

```python
import pandas as pd
from IPython.display import display
import warnings  
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import numpy as np
from random import *
import scipy.interpolate as sc

df = pd.read_csv('nba2k-full.csv')
df.head(10)
```

Полученный датасет будем хранить в переменной *df*. Проверим полученные данные на наличие пропусков и дубликатов, а также выведем общую информацию о таблице.

```python
print('Общая информация:')
df.info()
print()
print('Наличие дубликатов:')
print(df.duplicated().sum())
print()
print('Наличие пропусков:')
print(df.isnull().sum())
```

Первичный анализ не показал наличие дубликатов, но есть пропуски. Они присутствуют в столбцах *team* и *college*. Для того, чтоб не исказить данные, заменим пропуски на not specified. 

```python
col = ['team', 'college']

for i in col:
    df[i] = df[i].fillna('not specified')
```

Есть столбцы, которые представлены в неправильном типе данных, это тоже стоит исправить. 

Столбец *b_day*  необходимо перевести в фрмат даты

```python
for i in range(df.shape[0]):
    s = df['b_day'][i]
    if s[6] == '0' or s[6] == '1' or s[6] == '2':
        s = s[0:6] + '20' + s[6:8]
    else:
        s = s[0:6] + '19' + s[6:8]
    df['b_day'][i] = s
    s = ''
df['b_day'] = pd.to_datetime(df['b_day'], format='%m/%d/%Y')
```

Столбцы *height*, *weight* и *salary* наобходимо обработать и перевести в численный тип.

Для этого создадим новые столбцы:  *height_meters*, *weight_kg* и *salary_dollars*, которые будут содержать в себе рост в метрах, вес в килограммах, а зарплату в долларах. 

Такие преобразования необходимы для дальнейшей работы с данными. 

```python
df['height_meters'] = float(0)  #обработка роста
for i in range(len(df['height'])):
    s = df['height'][i].split('/')
    df['height_meters'][i] = float(s[1])
    s = ''

df['weight_kg'] = float(0)  #обработка веса
for i in range(len(df['weight'])):
    s = df['weight'][i].split('/')
    s1 = s[1].split()
    df['weight_kg'][i] = float(s1[0])
    s = ''
    s1 = ''

df['salary_dollars'] = int(0)  #обработка зарплаты
for i in range(len(df['salary'])):
    s = df['salary'][i]
    df['salary_dollars'][i] = int(s[1:])
    s = ''
```

## Интерполяция кривой

Сгенерируем от 5 до 20 точек со случайными разными координатами (x, y)

```python
cnt = randint(5, 20)

X = [random() * 10 for _ in range(cnt)]
Y = [random() * 10 for _ in range(cnt)]

print(X)
print(Y)
```

С помощью библиотеки scipy применим один из её алгоритмов интерполяции к сгенерированным данным:

```python
f1 = sc.interp1d(X, Y, kind='cubic')
lin = np.linspace(min(X), max(X), num=1000)

plt.plot(X, Y, 'o', lin, f1(lin))
plt.show()
```

Реализация интерполяционного многочлена Лагранжа.

```python
def lagran(x, y, a):
    ans = 0.0
    for i in range(len(y)):
        t = y[i]
        for j in range(len(y)):
            if i != j:
                t *= (a - x[j]) / (x[i] - x[j])
        ans += t
    return ans
```

Сравним полученную функцию со встроенной путем создания двух графиков:

```python
x_new = np.linspace(np.min(X),np.max(X),100)
y_new = [lagran(X,Y,i) for i in x_new]
plt.plot(X,Y,'o',x_new,y_new)
plt.title('Lagran')
plt.show()

f2 = sc.lagrange(X, Y)
plt.plot(X, Y, 'o', x_new, f2(x_new))
plt.title('Scipy')
plt.show()
```

Полученные графики идентичны. 

## Быстрая гистограмма

Построим гистограмму с помощью plt.hist

```python
a = [randint(0, 100) for i in range(randint(50, 100))]
print(a)
plt.hist(a, bins=len(set(a)))
```

Напишем свою реализацию разложения по контейнерам. Посчитаем максимальный, минимальный элементы, расстояние между колонками, и количество каждого элемента:

```python
from typing import List, Tuple, Union

def fast_hist(array: List[Union[int, float]],
              bins: int) -> Tuple[List[int], List[float]]:
    max_el = max(a)
    min_el = min(a)
    delta = (max_el - min_el) / bins
    bins_names = np.arange(min_el, max_el, delta)
    count_in_bin = np.zeros(bins)
    for el in array:
        count_in_bin[min(int((el - min_el) / delta), bins - 1)] += 1
    return (np.array([int(x) for x in count_in_bin]), bins_names)
```

Построим быструю гистограмму:

```python
value, bins = fast_hist(a, len(set(a)))

plt.bar(bins, value)
```

Полученная гистограмма идентична той, что была получена ранее  

Сравним скорость построения распределений: 

```python
%timeit plt.hist(a, len(set(a)))
```

```python
%timeit fast_hist(a, len(set(a)))
```
